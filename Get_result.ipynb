```python
import logging
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Suppress logging from Selenium and ChromeDriver
logging.basicConfig(level=logging.WARNING)  # Set logging level to WARNING to suppress less important logs
logger = logging.getLogger('selenium')
logger.setLevel(logging.WARNING)  # Only show warnings or errors

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run Chrome in headless mode
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--log-level=3")  # 3 = ERROR level

# Initialize the WebDriver
driver = webdriver.Chrome(options=chrome_options)

def check_download_link(prn, seat_start, seat_end):
    base_url = "https://siuexam.siu.edu.in/forms/resultview.html"  # Replace with your results page URL

    for seat_number in range(seat_start, seat_end + 1):
        try:
            driver.get(base_url)  # Open the results page

            # Wait for the PRN input field and enter the PRN
            try:
                prn_field = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.ID, "login"))  # Ensure the correct ID for the PRN input field
                )
                prn_field.clear()
                prn_field.send_keys(prn)
            except Exception as e:
                print(f"PRN input field not found: {e}")
                continue

            # Click the login button
            try:
                login_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.ID, "lgnbtn"))  # Ensure the correct ID for the login button
                )
                login_button.click()
            except Exception as e:
                print(f"Login button not clickable: {e}")
                continue

            time.sleep(2)  # Wait for the page to load

            # Attempt to find the seat number input field dynamically using the `name="txt"` attribute
            try:
                seat_field = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, f"//input[@name='txt']"))
                )
                seat_field.clear()
                seat_field.send_keys(str(seat_number))
            except Exception as e:
                print(f"Seat input field for {seat_number} not found: {e}")
                continue

            # Click the "View" button
            try:
                view_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, "//input[@value='View']"))
                )
                view_button.click()
            except Exception as e:
                print(f"View button not clickable for Seat Number {seat_number}: {e}")
                continue

            # Check for the download link
            try:
                download_link = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.XPATH, "//a[@class='btn btn-success btndef']"))
                )
                link_href = download_link.get_attribute("href")
                print(f"Download link available for Seat Number {seat_number}: {link_href}")
                break  # Exit the loop once a download link is found
            except Exception as e:
                print(f"No download link for Seat Number {seat_number}: {e}")

        except Exception as e:
            print(f"Error processing Seat Number {seat_number}: {e}")

# User input
prn = input("Enter your PRN: ")
seat_start = int(input("Enter the starting seat number: "))
seat_end = int(input("Enter the ending seat number: "))

# Run the function
check_download_link(prn, seat_start, seat_end)

# Quit the driver
driver.quit()
```
